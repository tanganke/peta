```python
def error_rate(y_true, y_pred):
    error = 0
    count = 0
    for i in range(len(y_true)):
        if y_true[i] != y_pred[i]:
            error += 1
        count += 1
    return error / count


@torch.no_grad()
def gather_predicts(model: nn.Module, dataloader: DataLoader):
    predictions = []
    for batch in dataloader:
        outputs = model.generate(batch["input_ids"])
        output_text = tokenizer.batch_decode(outputs, skip_special_tokens=True)
        predictions.extend(output_text)
    return predictions
```

```python
torch.set_float32_matmul_precision("medium")
task_0 = "glue-mrpc"
task_1 = "glue-rte"

n = 21
result = np.zeros((n, n), dtype=np.float32)

predicts_0 = {}
for lambda_0 in tqdm(np.linspace(-2, 2, n)):
    model: nn.Module = deepcopy(lora_pretrained_model)
    model.load_state_dict(
        state_dict_add(
            model.state_dict(),
            state_dict_mul(lora_task_vector[task_0], lambda_0),
            strict=False,
        ),
        strict=False,
    )
    model = fabric.setup_module(model)
    model.eval()
    predicts_0[lambda_0] = gather_predicts(model, val_loaders[task_0])

predicts_1 = {}
for lambda_1 in tqdm(np.linspace(-2, 2, n)):
    model: nn.Module = deepcopy(lora_pretrained_model)
    model.load_state_dict(
        state_dict_add(
            model.state_dict(),
            state_dict_mul(lora_task_vector[task_1], lambda_1),
            strict=False,
        ),
        strict=False,
    )
    model = fabric.setup_module(model)
    model.eval()
    predicts_1[lambda_1] = gather_predicts(model, val_loaders[task_1])

for i, lambda_0 in enumerate(tqdm(np.linspace(-2, 2, n))):
    for j, lambda_1 in enumerate(np.linspace(-2, 2, n)):
        model: nn.Module = deepcopy(lora_pretrained_model)
        model.load_state_dict(
            state_dict_add(
                model.state_dict(),
                state_dict_add(
                    state_dict_mul(lora_task_vector[task_0], lambda_0),
                    state_dict_mul(lora_task_vector[task_1], lambda_1),
                ),
                strict=False,
            ),
            strict=False,
        )
        model = fabric.setup_module(model)
        model.eval()

        predicts_on_0 = gather_predicts(model, val_loaders[task_0])
        predicts_on_1 = gather_predicts(model, val_loaders[task_1])

        result[i, j] = error_rate(
            predicts_0[lambda_0] + predicts_1[lambda_1], predicts_on_0 + predicts_on_1
        )
```

```python
np.save(
    f"results/flan-t5-base/lora_weight_disentanglement-{task_0}-{task_1}.npy",
    result,
)
```

```python
def plot_weight_disentanglement(
    finetune_mode,
    task_1: str,
    task_2: str,
    fig=None,
    ax: plt.Axes = None,
    cbar: bool = False,
):
    result = np.load(
        f"results/flan-t5-base/{finetune_mode}_weight_disentanglement-{task_1}-{task_2}.npy"
    )

    if ax is None:
        fig, ax = plt.subplots(figsize=(5, 4))
    im = ax.imshow(
        result,
        extent=[-2, 2, -2, 2],
        origin="lower",
        cmap="YlGnBu",
        vmin=0,
        vmax=2,
    )
    # plot a square x: 0, 1 y: 0, 1, red dashed line, no fill color
    ax.plot([0, 1, 1, 0, 0], [0, 0, 1, 1, 0], "r--", linewidth=1)
    # plot a x at the origin
    ax.plot([0], [0], "rx", linewidth=1)

    finetune_mode = 'LoRA' if finetune_mode == 'lora' else 'L-LoRA'
    ax.set_title(f"{finetune_mode}: {task_1[5:].upper()} - {task_2[5:].upper()}")
    ax.set_aspect("equal")
    ax.set_xlabel("$\lambda_1$")
    ax.set_ylabel("$\lambda_2$")
    ax.set_xticks([-2, -1, 0, 1, 2])
    ax.set_yticks([-2, -1, 0, 1, 2])

    if cbar:
        fig.colorbar(im, ax=ax)
    fig.tight_layout()
    return ax
```
